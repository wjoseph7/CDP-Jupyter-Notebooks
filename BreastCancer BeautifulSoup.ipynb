```python
#necessary imports 
import requests
import pandas as pd
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup
```

```python
#Scrape data from url
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/wpbc.data'
html_page = requests.get(url)
soup = BeautifulSoup(html_page.content, 'html.parser')
print(soup.prettify())
```

```python
#Convert to string and split strings into new lines
soup_string = str(soup)
soup_string = soup_string.split('\n')
```

```python
#Create a list to hold the data
numCols = 35

data = []
for i in range(numCols):
    data.append([])
    
for line in soup_string:
    line = line.split(',')
    #The last line is an empty string which we don't want to append
    if line[0] == '':
        break
    for i in range(numCols):
        data[i].append(line[i])
    
print(data)
```

```python
#Create  column names
cols = [ "id" , "outcome" , "time" ]
#We only use 32 instead of 35 here because we named 3 of the columns
for i in range(32):
    cols.append(str(i))

print(cols)
```

```python
#Initialize pandas dataframe 
df = pd.DataFrame.from_records(data,cols)
print(df)
```
