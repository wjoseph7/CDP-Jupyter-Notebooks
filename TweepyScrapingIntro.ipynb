```python
import os
import sys
from tweepy import API
from tweepy import OAuthHandler
from tweepy import Cursor
```

```python
#Logs into twitter using our keys
def get_twitter_auth():
    try:
        consumer_key = os.environ['TWITTER_CONSUMER_KEY']
        consumer_secret = os.environ['TWITTER_CONSUMER_SECRET']
        access_token = os.environ['TWITTER_ACCESS_TOKEN']
        access_secret = os.environ['TWITTER_ACCESS_SECRET']
    except KeyError:
        sys.stderr.write("TWITTER_* environment variables not set\n")
        sys.exit(-1)
    auth = OAuthHandler(consumer_key, consumer_secret)
    auth.set_access_token(access_token, access_secret)
    return auth
```

```python
#returns the profile we're using to scrape twitter
def get_twitter_client():
    auth = get_twitter_auth()
    client = API(auth, wait_on_rate_limit=True)
    return client
```

```python
#Scrape your own status
client = get_twitter_client()
for status in Cursor(client.home_timeline).items(10):
    print(status.text)
```

```python
#Get tweets with #unitedAirlines
tweet_batch = Cursor(client.search, q="#unitedAIRLINES", count=100).items()
for tweet in tweet_batch:
    print(tweet.text)
```
