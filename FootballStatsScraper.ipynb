```python
#necessary imports
import requests
from bs4 import BeautifulSoup
```

```python
#Take a look at an example url
base_url = 'https://www.pro-football-reference.com'
players_a = base_url+'/players/A'
html_page = requests.get(players_a)
soup = BeautifulSoup(html_page.content, 'html.parser')
print(soup.prettify)
```

```python
#Let's look at the player's whose last names start with A
for a in soup.find_all('a', href=True):
    if a['href'].startswith('/players/A/'):
        print("Found player:", a['href'])
```

```python
#Let's look at Ken Anderson
KenAnderson = base_url + '/players/A/AndeKe00.htm'
html_page = requests.get(KenAnderson)
soup = BeautifulSoup(html_page.content, 'html.parser')
print(soup.prettify)
```

```python
#Let's create a class to store our summary statistics for each player and print it out in a line
class fbPlayer():
    def __init__(self, name, games, AV, Qbrec, cmp, yds, YA, TD, Int, FantPt):
        self.name = name
        self.games = games
        self.AV = AV
        self.QBrec = QBrec
        self.cmp = cmp
        self.yds = yds
        self.YA = YA
        self.TD = TD
        self.Int = Int
        self.FantPt = FantPt
        
    def __str__(self):
        str = ''
        for var in vars(self):
            str += ' ' + self.__getattribute__(var)  
        return str
```

```python
#Scrape summary table
name = soup.find('h1', attrs={'itemprop': 'name'}).text

def scrapeAttribute(soup, string):
    html = soup.find('h4',text=string)
    try:
        text = html.find_next_sibling("p").text
    except:
        text = 'NA'
    return text

games = scrapeAttribute(soup, "G")
AV = scrapeAttribute(soup, "AV")
QBrec = scrapeAttribute(soup, "QBrec")
cmp = scrapeAttribute(soup, "Cmp%")
yds = scrapeAttribute(soup, "Yds")
YA = scrapeAttribute(soup, "Y/A")
TD = scrapeAttribute(soup, "TD")
Int = scrapeAttribute(soup, "Int")
FantPt = scrapeAttribute(soup, "FantPt")

KA = fbPlayer(name, games, AV, QBrec, cmp, yds, YA, TD, Int, FantPt)
print(KA)
```

```python
#Let's define a function to do all this for us
def scrapePlayer(soup):
    name = soup.find('h1', attrs={'itemprop': 'name'}).text
    games = scrapeAttribute(soup, "G")
    AV = scrapeAttribute(soup, "AV")
    QBrec = scrapeAttribute(soup, "QBrec")
    cmp = scrapeAttribute(soup, "Cmp%")
    yds = scrapeAttribute(soup, "Yds")
    YA = scrapeAttribute(soup, "Y/A")
    TD = scrapeAttribute(soup, "TD")
    Int = scrapeAttribute(soup, "Int")
    FantPt = scrapeAttribute(soup, "FantPt")

    return fbPlayer(name, games, AV, QBrec, cmp, yds, YA, TD, Int, FantPt)
```

```python
#Iterate through players alphabetically
#Automate scraping and print to csv
import string
alphabet = list(string.ascii_uppercase)

f = open('FootballStats.csv','a')

for letter in alphabet:
    letterExtension = '/players/'+letter
    print(letterExtension)
    url = base_url + letterExtension
    html_page = requests.get(url)
    letterSoup = BeautifulSoup(html_page.content, 'html.parser')
    
    for i,a in enumerate(letterSoup.find_all('a', href=True)):
        #We have to limit the number of scrapes per letter or it will take too long
        if i >= 100:
            break
        if a['href'].startswith(letterExtension):
            player = a['href']
            url = base_url + player
            html_page = requests.get(url)
            playerSoup = BeautifulSoup(html_page.content, 'html.parser')
            
            f.write(str(scrapePlayer(playerSoup)))
            f.write('\n')
f.close()
```

```python
#print out top ten players by touch down

names = []
TD = []

with open('FootballStats.txt','r') as f:
    lines = f.readlines()
    for line in lines:
        line = line.split(' ')
        line = [word for word in line if word != '']
        if len(line) == 11:
            names.append(line[0]+" "+line[1])
            if line[8] == "NA":
                TD.append(0)
            else:
                try:
                    TD.append(int(line[8]))
                except:
                    print(line)
                    break

    
f.close()

zipped = sorted(zip(TD,names))

for i in range(1,11):
    print(zipped[-i])
```
