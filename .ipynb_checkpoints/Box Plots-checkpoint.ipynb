```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
#Create a pandas dataframe of yelp reviews
yelp = pd.read_csv('yelp.csv')

#Create text length attribute
yelp['text length'] = yelp['text'].apply(len)
```

```python
#using seaborn's FacetGrid function allows us to plot the length distributions per star rating side by side with ease
g = sb.FacetGrid(data=yelp, col='stars')
g.map(plt.hist, 'text length', bins=50)
```

```python
#or if you must use matplotlib

#Create subsets for each star rating
x1 = yelp[yelp['stars']==1]
x1 = x1['text length']
x2 = yelp[yelp['stars']==2]
x2 = x2['text length']
x3 = yelp[yelp['stars']==3]
x3 = x3['text length']
x4 = yelp[yelp['stars']==4]
x4 = x4['text length']
x5 = yelp[yelp['stars']==5]
x5 = x5['text length']

#f is the figure and ax# is the particular subplot
f, (ax1, ax2, ax3, ax4, ax5) = plt.subplots(1, 5, sharey=True)
ax1.hist(x1,bins=50)
ax1.set_title('stars = 1')
ax2.hist(x2,bins=50)
ax2.set_title('stars = 2')
ax3.hist(x3,bins=50)
ax3.set_title('stars = 3')
ax4.hist(x4,bins=50)
ax4.set_title('stars = 4')
ax5.hist(x5,bins=50)
ax5.set_title('stars = 5')

#otherwise this will be too big for the Jupyter Notebook
f.set_size_inches(18.5, 3)
```

```python
#Let's look at a boxplot of the same data
sb.boxplot(x='stars', y='text length', data=yelp)
```

```python
#matplotlib code again
plt.boxplot([x1,x2,x3,x4,x5])
plt.xlabel("stars")
plt.ylabel("text length")
```
