```python
import networkx as nx #This is python's graph library
import matplotlib.pyplot as plt
```

Our goal is to visualize the graph of the Karate club network. The default visualization uses the Spring Layout. The Spring Layout teaches the edges between nodes like springs with an equilibrium length of 1/sqrt(N) where N is the total number of nodes. As you'll see this has nice visualization properties.

We want to plot the nodes with the core-nodes colored red.

```python
o='orange'
colors = ['r','r','r',o,o,o,o,o,o,o,o,o,o,o,o,o,o,o,o,o,o,o,o,o,o,o,o,o,o,o,o,o,'r','r']
G = nx.karate_club_graph() #This creates our Karate Club graph which is stored in networkx by default
nx.draw(G,with_labels=True,node_color=colors) #default visualization uses Spring Layout.
limits=plt.axis('off') # turn of axis: Honestly I'm not sure why this is necessary, but if I don't include it I get a warning.
```

We can compare the visualization to the circular graph shown below. Personally I prefer the Spring Layout because  it looks more natural. However, the circular plot will allow you to more easily compare different networks.

```python
nx.draw_circular(G,with_labels=True,node_color=colors)
```
